## lebel Encoding

```python
#label encoding is used for output column only

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Gender']= le.fit_transform(df['Gender'])
df['Educational_Status']= le.fit_transform(df['Educational_Status'])
df['Professional_Status']= le.fit_transform(df['Professional_Status'])
df['Marital_Status']= le.fit_transform(df['Marital_Status'])
df['Location']= le.fit_transform(df['Location'])
df['Live_With']= le.fit_transform(df['Live_With'])
df['Debt']= le.fit_transform(df['Debt'])
df['Exercise']= le.fit_transform(df['Exercise'])
df['Smoke']= le.fit_transform(df['Smoke'])
df['Illness']= le.fit_transform(df['Illness'])
df['Sleep_Time']= le.fit_transform(df['Sleep_Time'])
df['Work_Pressure']= le.fit_transform(df['Work_Pressure'])
df['Eating_Disorder']= le.fit_transform(df['Eating_Disorder'])
df['Cheated']= le.fit_transform(df['Cheated'])
df['Mistreated']= le.fit_transform(df['Mistreated'])
df['Suicidal']= le.fit_transform(df['Suicidal'])
df['Lost']= le.fit_transform(df['Lost'])
df['Prayer']= le.fit_transform(df['Prayer'])
df['Depressed']= le.fit_transform(df['Depressed'])
df['Drink']= le.fit_transform(df['Drink'])
```

```python
le.classes_
```

```python
df.head()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
df.corr()
```

```python
#drop one feature
df=df.drop(['Educational_Status'], axis = 1)
```

```python
df.shape
```

```python
#deside the input feature
x = df.drop(['Depressed'], axis = 1)
```

```python
x.head()
```

```python
#deside the ouput feature
y = df['Depressed']
```

```python
y
```

```python
#train test split from the dataset
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=.3,random_state=1)
```

## Logistic regration

```python
#import logistic regration algorithm
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(xtrain,ytrain)
```

```python
logmodel.score(xtest,ytest)
```

```python
predictions = logmodel.predict(xtest)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report
```

```python
ytest.shape
```

```python
print(classification_report(ytest,predictions))
```

## random forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=5)
rf.fit(xtrain,ytrain)
```

```python
pred6 = rf.predict(xtest)
```

```python
print(rf.score(xtest,ytest))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(ytest,pred6))
print(classification_report(ytest,pred6))
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(xtrain,ytrain)
```

```python
pred4 = knn.predict(xtest)
```

```python
print(knn.score(xtest,ytest))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(ytest,pred4))
print(classification_report(ytest,pred4))
```

## DT

```python
#Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier() #training
dtree.fit(xtrain,ytrain)
```

```python
pred3 = dtree.predict(xtest)
```

```python
print(dtree.score(xtest,ytest))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(ytest,pred3))
print(classification_report(ytest,pred3))
```
